```python
# Fix Missing Values
# Drop rows where Crop name is missing (only 9 rows — safe to drop)
apy_df = apy_df.dropna(subset=['Crop'])
```

```python
# Fill missing Production with median
apy_df['Production'] = apy_df['Production'].fillna(apy_df['Production'].median())
```

```python
# Confirm
print("Nulls remaining:")
print(apy_df.isnull().sum())
```

Remove Outliers from Yield ⚠️


```python
# Anything too far above Q3 or too far below Q1 is considered an outlier.
# Calculate Q1, Q3 and IQR for Yield
Q1 = apy_df['Yield'].quantile(0.25)   # 25th percentile
Q3 = apy_df['Yield'].quantile(0.75)   # 75th percentile
IQR = Q3 - Q1

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")

# Define outlier boundaries
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Keeping Yield between {lower:.2f} and {upper:.2f}")

# Keep only rows within boundaries
apy_df = apy_df[(apy_df['Yield'] >= lower) & (apy_df['Yield'] <= upper)]

print("APY shape after removing outliers:", apy_df.shape)
```

```python
# Feature Engineering 🔢
# Feature Engineering means preparing your data so the ML model can understand it.
# Models only understand numbers — they can't read words like "Kharif" or "Karnataka". 
# So we need to convert text columns into numbers.
```

APY Dataset
State    → text ❌  needs conversion
District → text ❌  needs conversion  
Crop     → text ❌  needs conversion
Season   → text ❌  needs conversion

Crop Dataset
label    → text ❌  needs conversion
N,P,K etc → already numbers ✅


Two Ways to Encode Text
Method 1 — Label Encoding
Converts each category to a number
Kharif      → 0
Rabi        → 1
Summer      → 2
Whole Year  → 3

Use when: Column has many unique values like State, District, Crop
Method 2 — One Hot Encoding
Creates a new column for each category with 0 or 1
Season_Kharif  Season_Rabi  Season_Summer
     1               0            0        ← this row is Kharif
     0               1            0        ← this row is Rabi

Use when: Column has few unique values like Season (only 6 types)

Here we'll use Label Encoding for everything
It's simpler and works well with Random Forest (the model we'll use in Step 6)

```python
print(apy_df.columns.tolist())
```

```python
# Remove any leading/trailing spaces from ALL column names
apy_df.columns = apy_df.columns.str.strip()

# Confirm
print(apy_df.columns.tolist())
```

```python
 # Encode APY Dataset
le = LabelEncoder()

# Encode each text column
apy_df['State']    = le.fit_transform(apy_df['State'])
apy_df['District'] = le.fit_transform(apy_df['District'])
apy_df['Crop']     = le.fit_transform(apy_df['Crop'])
apy_df['Season']   = le.fit_transform(apy_df['Season'])

# Check the result
print(apy_df.head())
print()
print(apy_df.dtypes)
```

```python
 # Encode Crop Dataset
crop_df['label'] = le.fit_transform(crop_df['label'])

print(crop_df.head())
print()
print(crop_df.dtypes)
```

```python
# Split Features and Target
# APY Dataset
X_apy = apy_df.drop('Yield', axis=1)
y_apy = apy_df['Yield']

# Crop Dataset
X_crop = crop_df.drop('label', axis=1)
y_crop = crop_df['label']

print("APY Features shape:", X_apy.shape)
print("APY Target shape:", y_apy.shape)
print("Crop Features shape:", X_crop.shape)
print("Crop Target shape:", y_crop.shape)
```

```python
# Train Test Split

X_apy_train, X_apy_test, y_apy_train, y_apy_test = train_test_split(
    X_apy, y_apy, test_size=0.2, random_state=42)

X_crop_train, X_crop_test, y_crop_train, y_crop_test = train_test_split(
    X_crop, y_crop, test_size=0.2, random_state=42)

print("APY Training set:", X_apy_train.shape)
print("APY Testing set:", X_apy_test.shape)
print()
print("Crop Training set:", X_crop_train.shape)
print("Crop Testing set:", X_crop_test.shape)
```

```python
 # Building Crop Recommendation Model
# Create the model
# n_estimators = number of trees (100 is a good default)
# random_state = so results are same every time you run
crop_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
# This is where the magic happens — model learns from training data
print("Training Crop model... please wait ⏳")
crop_model.fit(X_crop_train, y_crop_train)
print("Done! ✅")

# Test the model
y_crop_pred = crop_model.predict(X_crop_test)

# Check accuracy
accuracy = accuracy_score(y_crop_test, y_crop_pred)
print(f"Crop Recommendation Accuracy: {accuracy * 100:.2f}%")
```

```python
# Building Yield Prediction Model

# Create the model
# Regressor because we're predicting a NUMBER (yield)
# Classifier is for predicting a CATEGORY (crop name)
yield_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
print("Training Yield model... please wait ⏳")
print("(This may take 1-2 minutes — large dataset!)")
yield_model.fit(X_apy_train, y_apy_train)
print("Done! ✅")

# Test the model
y_apy_pred = yield_model.predict(X_apy_test)

# Check performance
rmse = np.sqrt(mean_squared_error(y_apy_test, y_apy_pred))
r2   = r2_score(y_apy_test, y_apy_pred)

print(f"Yield Prediction RMSE: {rmse:.4f}")
print(f"Yield Prediction R²  : {r2:.4f}")
```

```python
# Check training vs testing performance(Overfitting)
train_pred = yield_model.predict(X_apy_train)

train_r2   = r2_score(y_apy_train, train_pred)
test_r2    = r2_score(y_apy_test, y_apy_pred)

train_rmse = np.sqrt(mean_squared_error(y_apy_train, train_pred))
test_rmse  = np.sqrt(mean_squared_error(y_apy_test, y_apy_pred))

print(f"Training R²  : {train_r2:.4f}")
print(f"Testing  R²  : {test_r2:.4f}")
print(f"Difference   : {(train_r2 - test_r2):.4f}")
print()
print(f"Training RMSE: {train_rmse:.4f}")
print(f"Testing  RMSE: {test_rmse:.4f}")
```

```python
# Model Evaluation 📊

# Crop Model Evaluation(Confusion Matrix)
# Generate confusion matrix
cm = confusion_matrix(y_crop_test, y_crop_pred)

# Plot it
plt.figure(figsize=(15, 12))
sns.heatmap(cm, 
            annot=True,        # show numbers inside boxes
            fmt='d',           # show as integers
            cmap='Blues',      # color scheme
            xticklabels=le.classes_,   # crop names on x axis
            yticklabels=le.classes_)   # crop names on y axis

plt.title('Confusion Matrix — Crop Recommendation Model')
plt.xlabel('Predicted Crop')
plt.ylabel('Actual Crop')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Classification Report
# Detailed report for each crop
print("Classification Report:")
print(classification_report(y_crop_test, y_crop_pred, 
                            target_names=le.classes_))
```

```python
# Feature Importance for Crop Model
# Get importance scores from model
feature_names = X_crop.columns
importances   = crop_model.feature_importances_

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=importances, y=feature_names, 
            palette='viridis',
            order=feature_names[importances.argsort()[::-1]])
plt.title('Feature Importance — Crop Recommendation')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()
```
